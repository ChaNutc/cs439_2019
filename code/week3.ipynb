# Masking

```python
import numpy as np #pip3 install scipy
x=np.arange(0,6*np.pi,0.1)
y=np.sin(x)
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.plot(x,y,'.b')
```

```python
x=np.arange(0,6*np.pi,0.1)
x.shape
```

```python
y=np.sin(x)+np.random.randn(x.shape[0])*0.5
```

```python
%matplotlib inline
from matplotlib import pyplot as plt #pip3 install matplotlib
plt.style.use('ggplot')
plt.plot(x,y,'.b')
```

```python
y=np.sin(x)+np.random.randn(x.shape[0])*0.5
%matplotlib inline 
from matplotlib import pyplot as plt
plt.style.use('ggplot')
for i in range(len(y)):
    if y[i]<0:
        y[i]=0.0
plt.plot(x,y,'b')
```

```python
plt.plot(x,y,'.r')
```

# Masking

```python
y[y<0]=0
```

```python
plt.plot(x,y,'.b')
```
