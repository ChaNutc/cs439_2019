# Regression

# Gradient Descent
$\theta_{t+1} = \theta_{t+1} - \alpha. \frac{\partial L}{\partial\theta}$

```python
import numpy as np
```

```python
x=np.array([1,2,3])
yo=np.array([3,9,8])
```

```python
def yp(x,m,c):
    return m*x+c
```

```python
yp(x,m=1,c=2)
```

```python
yp(x,m=1,c=2)-yo
```

```python
np.sum(yp(x,m=1,c=2)-yo)
```

```python
np.sum((yp(x,m=1,c=2)-yo)*x)
```

```python
m=1
c=2
alpha = 0.1
for i in range(10):
    m= m - alpha*np.sum((yp(x,m,c)-yo)*x)
    c= c - alpha*np.sum((yp(x,m,c)-yo))
    print(m,c)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
mx=np.arange(0,4,1)
my=yp(mx,m,c)
plt.plot(mx,my)
```

```python
m=1
c=2
alpha = 0.1
for i in range(10):
    m= m - alpha*np.sum((yp(x,m,c)-yo)*x)
    c= c - alpha*np.sum((yp(x,m,c)-yo))
    my=yp(mx,m,c)
    plt.plot(x,yo,'.k')
    plt.plot(mx,my,color = (1.0-i/10,i/10,0.0))
```

```python
M=100
x=np.random.rand(100)*3
yo=yp(x,m=3,c=10)+np.random.randn(100)*10
plt.plot(x,yo,'.k')
```

```python
m=1
c=-10
N=500
alpha = 0.1
for i in range(N):
    m= m - alpha*np.sum((yp(x,m,c)-yo)*x)/M
    c= c - alpha*np.sum((yp(x,m,c)-yo))/M
    my=yp(mx,m,c)
    plt.plot(x,yo,'.k')
    plt.plot(mx,my,color = (1.0-i/N,i/N,0.0))
```
